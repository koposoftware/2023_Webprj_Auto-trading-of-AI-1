```python
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
from pykrx import stock as st
import pickle
import plotly.express as px
import plotly.graph_objects as go
```

### csv 불러오기 실습

DataFrame : 2차원 데이터 -> table

```python
date_label = pickle.load(open('date_label.sav', 'rb'))
stock_label = pickle.load(open('stock_label.sav', 'rb'))
adjusted = pd.read_csv("액분.csv").reset_index(drop=True)
```

```python
adjusted['isin'] = adjusted['isin'].str.replace(')','').str.split('(').str[1]
adjusted = pd.merge(adjusted,stock_label,how='inner')
adjusted = adjusted['isin'].unique().tolist()
```

```python
a = adjusted[:100]
b = adjusted[100:200]
c = adjusted[200:300]
d = adjusted[300:]
ohlcv = pd.DataFrame([])
```

```python
yf.pdr_override()

for tickers in [a,b,c,d]:
    for ticker in tickers:
        ticker = ticker + '.KS'
        tmp1 = pdr.get_data_yahoo(ticker, start='2013-01-01', end='2023-07-18').reset_index()
        tmp1['isin'] = ticker.replace('.KS','')
        ohlcv = pd.concat([ohlcv, tmp1])

ohlcv[['Open', 'High', 'Low', 'Adj Close']] = ohlcv[['Open', 'High', 'Low', 'Adj Close']].astype(int)
ohlcv = ohlcv[['isin', 'Date', 'Open', 'High', 'Low', 'Adj Close']]
ohlcv = ohlcv.rename({'Date':'s_date','Adj Close':'close'}, axis = 1)
ohlcv.to_csv('ohlcv최종.csv', sep=',', index = False)
```
